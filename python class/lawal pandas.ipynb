```python
# use normal function to create new columns
def create(x):
    if x == 'Iris-setosa':
        x = 1
    elif x == 'Iris-virginica':
        x = 2
    return x


df['new_column'] = df['species'].apply(create)
```

```python
df.sample()
```

```python
# use lambda function to create new columns
df['column2'] = df['species'].apply(lambda x: 1 if x=='Iris-setosa' else 2 if x == 'Iris versicolor' else 3)
df
```

```python
df['new'] = df['new'].apply(lambda x: x * 20)
```
